```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np

master_df = pd.read_csv('data/cbb.csv')

default_training_factors = [
    'Win_Pct',
    'ADJOE',
    'ADJDE',
    'BARTHAG',
    'EFG_O',
    'EFG_D',
    'TOR',
    'TORD',
    'ORB',
    'DRB',
    'FTR',
    'FTRD',
    '2P_O',
    '2P_D',
    '3P_O',
    '3P_D',
    'ADJ_T',
    'WAB'
]

# scaler = MinMaxScaler()
# master_df[default_training_factors] = scaler.fit_transform(master_df[default_training_factors])
# print(master_df.head())

def preprocess_historic_data(df, training_factors=default_training_factors):
    """
    Convert the df from the csv file to the df ready for learning.
    
    Should accomplish several things:
        - Filter to only use the relevant columns
        - Scale factors to prepare for KNN
        - Create any additional columns needed in DataFrame (ex. Win_Pct)
    """
    # Create any additional columns needed in DataFrame
    df['Win_Pct'] = df.apply(lambda row: row.W / row.G, axis=1)

    # Scale factors to prepare for KNN
    scaler = MinMaxScaler()
    df[training_factors] = scaler.fit_transform(df[training_factors])
    
    # Filter to only use the relevant columns
    df = df[training_factors]
    
    return df

print(master_df.head())

all_teams_scaled_and_vectorized_df = preprocess_historic_data(master_df)
all_teams_scaled_and_vectorized_df.head()

# Be good to get the indices which match a query, can extend this to serve several use cases
```

```python
df23 = pd.read_csv('data/cbb23.csv')
df23 = df23.drop(columns=['Rk'])

df23.head()
```

```python
import re

# Many cells have 2 values entered on seperate lines. Invariably, we only want the first
for i, row in df23.iterrows():
    for j, cell in row.items():
        if '\n' in cell:
            df23.iloc[i][j] = re.split('\n', cell)[0]

print(df23)
        
        
```

```python
# Other cleaning actions required for current season data:

# Remove rows that are duplicates of the title row
rows_to_drop = []
for i, row in df23.iterrows():
    if row['2P%D'] == '2P%D':
        rows_to_drop.append(i)
        
df23 = df23.drop(rows_to_drop)
print(df23)
```

```python
# This data has Rec(ord) column, we need W(ins)
df23['W'] = df23.apply(lambda row: re.split('-', row['Rec'])[0], axis=1)

df23.head()
                       
```

```python
df23['YEAR'] = df23.apply(lambda row: 2023, axis=1)

df23.head()
```

```python
# Sync capitalization of column labels with master_df
old_new_col_label = {
    'Team': 'TEAM',
    'Conf': 'CONF',
    'AdjOE': 'ADJOE',
    'AdjDE': 'ADJDE',
    'Barthag': 'BARTHAG',
    'EFG%': 'EFG_O',
    'EFGD%': 'EFG_D',
    '2P%': '2P_O',
    '2P%D': '2P_D',
    '3P%': '3P_O',
    '3P%D': '3P_D',
    'Adj T.': 'ADJ_T'
}

df23 = df23.rename(columns=old_new_col_label)

df23.head()
```

```python
from sklearn.model_selection import train_test_split

historic_teams_query = master_df.query('YEAR != 2023')
historic_teams_df = master_df.iloc[historic_teams_query.index.values]
historic_teams_df.head()
```

```python
X = historic_teams_scaled_and_vectorized_df.to_numpy()
Y = historic_teams_df.apply(lambda row: 0 if pd.isna(row['SEED']) else 1, axis=1).to_numpy()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8)
```

```python
# Use existing default SVC for makes_tournament_classifier and collect some metrics

makes_tournament_classifier = svm.SVC()
makes_tournament_classifier.fit(X_train, Y_train)
```

```python
Y_pred = makes_tournament_classifier.predict(X_test)
print(Y_pred)
```

```python
from sklearn.metrics import accuracy_score

# Baseline accuracy score (Pre GridSearchCV)
print(accuracy_score(Y_test, Y_pred))
```

```python
# Run GridSearchCV for makes tournament predictor
from sklearn.model_selection import GridSearchCV

base_classifier = svm.SVC()

params = {
    'C': [1, 3, 5, 7, 9],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'shrinking': [True, False],
    'decision_function_shape': ['ovo', 'ovr'],
    'gamma': ['scale', 'auto'],
    'coef0': [1, 3, 5, 7, 9],
    'probability': [True, False]
}

enhanced_classifier = GridSearchCV(base_classifier, params)
enhanced_classifier.fit(X_train, Y_train)
```

```python
Y_pred = enhanced_classifier.predict(X_test)

print(accuracy_score(Y_test, Y_pred))
```
